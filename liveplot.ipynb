```python
import pandas as pd
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
CSV_FILE = "database/data.csv"

while True:
    try:
        df = pd.read_csv(CSV_FILE)

        clear_output(wait=True)

        plt.figure(figsize=(10, 5))
        plt.plot(df["final_weight"])
        plt.title("Liveplot: Endgewicht")
        plt.xlabel("Flasche")
        plt.ylabel("Final Weight [g]")
        plt.grid(True)
        plt.show()

        time.sleep(5)

    except FileNotFoundError:
        clear_output(wait=True)
        print("CSV-Datei noch nicht gefunden...")
        time.sleep(5)

    except Exception as e:
        clear_output(wait=True)
        print("Fehler:", e)
        time.sleep(5)
```
